```python
# Important Libraries
import lasio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
```

```python
def las_read(file):
    las = lasio.read(file)
    header = las.header
    data = las.df()
    data.reset_index(inplace = True)
    for curve in las.curves:
        print(curve.mnemonic,"------", curve.unit,"------", curve.descr,"------", len(curve.data))
    return data, header
```

```python
df = df2.dropna()
df.reset_index(inplace=True)
df=df.drop(columns =['index'])
```

```python
cols = ['DRY_WEIGHT_HEMATITE' ,'DRY_WEIGHT_ILLITE' , 'DRY_WEIGHT_KAOLINITE','DRY_WEIGHT_CHLORITE',
        'DRY_WEIGHT_QUARTZ','DRY_WEIGHT_CALCITE','DRY_WEIGHT_DOLOMITE','DRY_WEIGHT_K-FELDSPAR','UWATER']

df[cols]
```

```python
colors = ['#33B8FF', '#e6beff', '#ffe119', 'red', '#f58231', '#911eb4', 'grey', '#f032e6', '#bcf60c', '#0E4E70',
            '#008080', '#e6beff', '#9a6324', '#fffac8', '#200201', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080']

cols2 = ['BOUND_WATER','CALCITE','CHLORITE','DOLOMITE','HEMATITE',
         'ILLITE','K-FELDSPAR','KAOLINITE','QUARTZ','UWATER']

def depth_vs_minerals(df, cols):
    depth = df['MD']
    fig, ax = plt.subplots(figsize=(16, 50))
    
    minerals = df[cols]
    
    cmap = ListedColormap(colors)
    
    baseline = np.zeros_like(depth)
    
    for idx, mineral in enumerate(minerals):
        mineral_values = df[mineral]
        ax.fill_betweenx(depth, baseline, baseline + mineral_values, color=cmap(idx), label=mineral)
        baseline += mineral_values

    ax.set_xlabel('Mineral Concentration', fontweight='bold')
    ax.set_ylabel('Depth (MD)', fontweight='bold')
    ax.set_title('Mineral Concentration Profile Along Depth', fontweight='bold')
    
    ax.invert_yaxis()
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.rcParams['font.size'] = 30 
    plt.rcParams['font.family'] = 'arial' 
    plt.show()
    
depth_vs_minerals(df, cols2)
```
